# А що таке класи та обєкти?

> В класичному вигляді для пояснення класі і обєкту говорять що: Клас це є шаблон, блупрінт (схема/креслення) який описує якесь фізичне явище, чи предмет, чи абстракцію яку ми хочем описати для компютера (для прикладу, пере ти як створити машину, малюють її концепт, створюють дизайн, інженери роблять креслення для того щоб описати деталі та як їх складати). По інструкції, яку надає Клас ми можемо створити Обєкт який є вираженням фізичної сутності того що описує клас. # Опишемо абстракцію предмета Меч, який буде використовуватись у грі

> спочатку спробуємо задати Меч якимось простішими елментами мови Пайтон (без використання ООП, принафмні в явному вигляді) Меч - це предмет, який може випасти гравцю, він має певну назву, в нього є характеристика рідкісності. Меч має якісь бойові характеристики і має якусь міцність/витривалість. І також особливі мечі мають якись спеціальний ефект.

```python
# ми можемо використати базовий ти даних Словник для опису меча
sword_0 = {"name": "Короткий меч", "rarity": "Green", "damage": int(5), "vitality": int(10), "bonus": None}

# задані характеристики будуть статичними/незмінними
def sword_info(sw:dict):
    print(f"""<<< Характеристика предмету {sw["name"]} >>>
Рідкісність: {sw["rarity"]}
Нанесення Шкоди: {sw["damage"]}
Витривалість: {sw["vitality"]}
Унікальний ефект: {sw["bonus"]}""")

sword_info(sword_0)

from random import randint, choice
# Закоментуємо ввід даних щоб могти виконати комірку
#my_value = input("Введіть число від 1 до 3:")
#value = randint(1, 3)
#if int(my_value) == value:
#    print("Вітаємо ви отримали Меч, його характеристики")
#    sword_info(sword)
#else:
#    print("Вам нічого не випало.")
```

> Якщо ми хочемо передбачити що даний тип меча може мати різні характеристики ми повинні ускладнити його опис;

> Давайте перепишемо представлення меча і зробимо деякі поля випадковими;

> Випадко вибрані поля ніяк не будуть повязувати в собі характеристики меча, бо кожен рандомний вибір буде незалежним

```python
sword_1 = {"name": "Короткий меч", 
         "rarity": choice(["White", "Blue", "Green", "Red", "Epic"]), 
         "damage": int(5), 
         "vitality": int(10), 
         "bonus": choice([None, "Вампіризм", "Вдача +2", "Покритий отрута"])}

sword_info(sword_1)
```

> у попередньому коді, кожен вибір є незалежним, давайте заробимо залежність всього меча від конкретного значення рідкісності;

> В залежності від рідкісності предмету будемо визначати його "стати";

```python
# Визначаємо мапу коефіцієнтів які будуть домножуватись на наші стати відповідно до рідкісності предмету
stats_map = {"White": 1, 
             "Blue": 2, 
             "Green": 3, 
             "Red": 5, 
             "Epic": 8, 
             "Legendary": 12}

bonus_map = {"White": None, 
             "Blue": "Вдача +2", 
             "Green": "Вампіризм", 
             "Red": "Покритий отрута", 
             "Epic": ["Вампіризм", "Парирування"], 
             "Legendary": ["Критичний удар +2", "Зцілення +1", "Розкидає бомби"]}

rariry = choice(list(stats_map.keys()))
print(f"Випала рідкісність: {rariry}")

print("Ми задонатили та отримали бонус для всього акаунта на 1% до дамагу!")
donate = 1.01

sword_2 = {"name": "Короткий меч", 
         "rarity": rariry, 
         "damage": round(5 * stats_map[rariry] * donate), 
         "vitality": int(10) * stats_map[rariry], 
         "bonus": bonus_map[rariry]}

sword_info(sword_2)

sword_3 = {"name": "Двохручний меч", 
         "rarity": rariry, 
         "damage": round(7 * stats_map[rariry] * donate), 
         "vitality": int(8) * stats_map[rariry], 
         "bonus": bonus_map[rariry]}

sword_info(sword_3)

print(f"Перший гравець з мечем 2 атакує гравця з мечем 3, Меч 3 залишилось витривалості: {sword_3['vitality'] - sword_2['damage']}")
```

> якщо дивитись по аналогіям, то опис Меча простими засобавими Пайтон буде виглядати як реляційні Бази даних - ми маючи Мапи з набором значень, по ключовому полю rarity витягнули цю інформацію та сформували на Меч.

> якщо у нас зявиться додаткова характеристика, нам прийдеться багато чого виправляти і деколи зміни можуть бути не сумісними. Може бути що Меч і його характеристики удуть залежати від декількох ключових полів, у цьому випадку в нас буде дуже багато індексів.

> опис Меча буде ставати тільки складнішим по мірі того як ми будемо додавати до нього функціонал, тому нам потрібно змінити підхід до створення/опису меча у програмі і це робиться за допомогою ООП

> За допомогою ООП ми можемо описати предмет і надати йому стільки характеристик стіки нам буде потрібно, довати ці харакеристити або змінюват їх не боячись сильно пошкодити роботу програми

> спробуємо створити пустий клас який просто буде демострувати роботу ООП;

> в такому класі всі його атрибути (характеристики або параметри обєкта) будуть динамічними і не будуть постійними для кожного з обєктів;

> Створивши пустий обєкт, він буде не зовсім пустим бо по замовчуванню завжди існують так звані Магічні методи (dundle methods) які можна викливати за допомогою команди . (за допомогою dot ми можемо доступитись до будь-чого всередині обєкта);

```python
class SwordEmpty:
    """Пустий клас що може бути прототипом меча із заданими динамічними атрибутами"""
    def info(self) -> None:
        """Це тепер Метод який виводить інформацію про Меч.
        :param self: Нам потрібен лише вказівник на обєкт з яким ми працюємо; 
        """
        print(f"""<<< Характеристика предмету {self.name} >>>
Рідкісність: {self.rarity}
Нанесення Шкоди: {self.damage}
Витривалість: {self.vitality}
Унікальний ефект: {self.bonus}""")

se = SwordEmpty()

print(f"Ми створили {SwordEmpty} та обєкт до нього: {se} або можеми викликати як {se.__str__()}")
```

> ми повинні повторити за допомогою ООП той самий функціонал що ми зробли для Меча. Меч повинен мати певні характеристики та відповідати певні якості та мати певні бонуси;

> ми говорили про атрибути обєкта - ними ми можемо оперувати і задавати характеристики меча;

```python
se.name = "Двохручний меч"
se.rarity = "White"
se.damage = 3
se.vitality = 10
se.bonus = ["Покритий отрута"]

print(f"Ми можемо подивитись що саме міститься тепер в нашому обєкті:\n{se.__dict__}")
print(f"Та можемо доступитись до нашин атрибутів за допомогою dot: {se.name} має атаку {se.damage}")

se.info()
```

> якщо охарактеризувати наш обєкт, то можна сказати що він представляє сукупність даних та методів які зібрані в одну сутність та представляють собою характеристику якогось фізичного явища, а саме Меча;

> якщо ми будемо створювати обєкти але не задавати цим обєктам потрібні нам динамічні атрибути, то функіонал нашого Класу/Обєкта може бути порушеним;

```python
# При роботі з динамічними атрибутаи, їх треба наперед задавати при ініціалізації кожного з обєктів
s1 = SwordEmpty()
s1.name = "Простий меч"
s1.rarity = "Green"
s1.damage = 4
s1.vitality = 8
s1.bonus = ["Зцілення +1"]
s1.info()

s2 = SwordEmpty()
# Наступний кусок коду видасть помилку `AttributeError: 'SwordEmpty' object has no attribute 'name'`
#s2.info()
```

> Для того щоб все працювало правильно, ми можемо наперед передбачити якими саме атрибутами буде володіти наш Обєкт та прописати їх у нашому класі;
> ми хочемо щоб функціонал класу працював завжди для кожного створеного обєкту;

```python
class SwordsFirst:
    """Це наш перший клас який використовує конструктор для визначення наперед заданих атрибутів"""
    def __init__(self, name: str, rarity: str, damage: int, vitality: int, bonus: list) -> None:
        """Це Конструктор, він ініціалізує Обєкт за допомогою Аргументів які передаються в нього та присвоюються Атрибутам
        
        :param name: Цей аргумент визначає назву Меча;
        :param rarity: Цей аргумент відповідає за рідкісність предмету Меча;
        :param damage: Характеристика нанесення шкоди нашим Мечем;
        :param vitality: Характеристика міцності Меча;
        :param bonus: Список додаткових бонусів для Меча;
        """
        self.name = name
        self.rarity = rarity if rarity in ["White", "Blue", "Green"] else None
        self.damage = damage
        self.vitality = vitality
        self.bonus = bonus
    
    @property
    def hit(self):
        """Визначає кількісь шкоди яку ми можемо нанести супротивнику"""
        return self.damage
    
    @property
    def rarity_type(self):
        """Дана проперті буде нам надавати інформацію про рідкісність предмету та визначати чи відповідає вона умовам"""
        rarity_map = {
            "White": "Предмет має Базову якість.",
            "Blue": "Предмет має Покращену якість та збільшені характеристики.",
            "Green": "Предмет має Продвинута якість та достатньо високі показники нанесення шкоди та витривалості."
        }
        if self.rarity in rarity_map.keys():
            return rarity_map[self.rarity]
        return "Предмет зломаний, можете викинути його у сміття."

    def info(self) -> None:
        """Це тепер Метод який виводить інформацію про Меч.
        :param self: Нам потрібен лише вказівник на обєкт з яким ми працюємо; 
        """
        print(f"""<<< Характеристика предмету {self.name} >>>
Рідкісність: {self.rarity}, {self.rarity_type}
Нанесення Шкоди: {self.damage}
Витривалість: {self.vitality}
Унікальний ефект: {self.bonus}""")
        
# Якщо ми вказуємо аргументи у правильному порядку то на не потрібно вказувати чому саме буде відповідати значення через кому
s1 = SwordsFirst("Простий меч", "Green", 4, 8, ["Зцілення +1"])
# також можна здавати аргументи в явному вигляді
s2 = SwordsFirst(name="Катана", rarity="Blue", damage=5, vitality=6, bonus=["Критичний удар +2"])

s3 = SwordsFirst("Надломаний меч", None, 1, 1, [])

# Тепер кожен створений обєкт буде мати базові атрибути які будуть спільними для всіх обєктів (бо ми прописали вигляд обєкту в конструкторі)
# також, всі обєкти повинні бути створені та ініціалізовані даними аргументами які відповідають конкретним атрибутам
# і в такому випадку, функціонал класу буде працювати з будь-яким обєктом
s1.info()
s2.info()
s3.info()
```

- деколи бувають ситуації коли потрібно мати поля які не можуть змінюватись, такими полями є property. води повнісю такі ж як атрибути але вони імунні;

# Методи в класах або дії що можуть виконувати обєкти
- атрибути та проперті - це просто інформаці/дані про обєкт, що несе певний інформаційних характер. Ми знаємо що з себе представляє це меч, які в нього характеристики, але це нічого не говорить про те як ним користоватись або на що він здатний;
- також меч може виконувати певні дії (атакувати або парирувати удари) або ці дії можуть виконуватись нади ним (накладати бафи або дебафи на меч);
- такі дії ще називаються методами, і залежності від того над чим ці дії виконуються бувають: методи (дії з обєктом), класові методи (альтернативні конструктори) та статичні методи (це прості функції які роблять якісь обчислення незалежні від обєкта);
- давайте модифікуємо клас SwordsFirst, переназвемо його як SwordsSecond та допишимо в нього методи якими буде володіти наш Меч;
- основі ми залишимо незмінно, конструктор та проперті будуть такими ж;

```python
class Buff:
    def __init__(self, obj) -> None:
        self.name = None
        self.damage = None
        self.vitality = None
        self.obj = obj

    def sharpening(self):
        """Заточення"""
        # Базові значення накладених бафів
        if self.obj.buff:
            return "Неможливо накласти 2 Бафи"
        
        self.name = "Заточення"
        self.damage = 4
        self.vitality = -1
        print(f"Заточення: Шкода {self.damage} ||| Витривалість {self.vitality}")
        self.obj.damage += self.damage
        self.obj.vitality += self.vitality
        return "Меч заточено"
    
    def poisoning(self):
        """Змазуємо Меч отрутою"""
        if self.obj.buff:
            return "Неможливо накласти 2 Бафи"
        self.name = "Отрута"
        self.damage = 2
        self.vitality = 0
        print("Отрута")
        self.obj.damage += self.damage
        return "Змазано отрутою"
    
    def enchantment(self):
        """Зачаровуємо Меч на витривалість"""
        if self.obj.buff:
            return "Неможливо накласти 2 Бафи"
        self.name = "Зачарування"
        self.damage = 0
        self.vitality = 5
        print("Зачарування")
        self.obj.vitality += 5
        return "Накладено чари"
    
    def crystals(self):
        """Застосовуємо криста Мани"""
        if self.obj.buff:
            return "Неможливо накласти 2 Бафи"
        self.name = "Кристал Мани"
        self.damage = -1
        self.vitality = 3
        print("Застосовуємо кристал мани")
        self.obj.vitality += 3
        self.obj.damage -= 1
        return "Застосовано кристал мани"

    def berserk(self):
        """Впадаємо в стан Берсерка"""
        if self.obj.buff:
            return "Неможливо накласти 2 Бафи"
          
        self.name = "Берсерк"
        self.damage = self.obj.damage * 2 # вдвічі збільшуємо значення шкоди 
        self.vitality = self.obj.vitality / 2
        print("Ввійшли в стан Берсерка")
        self.obj.vitality = self.vitality
        self.obj.damage = self.damage
        return "Ввійшли в стан Берсерка"


class SwordsSecond:
    """Це наш перший клас який використовує конструктор для визначення наперед заданих атрибутів"""
    # Це класова глобальна змінна, її можна використовувати будь-де всередині класі або ж передавати кудиінде
    rarity_map = {
            "White": "Предмет має Базову якість.",
            "Blue": "Предмет має Покращену якість та збільшені характеристики.",
            "Green": "Предмет має Продвинута якість та достатньо високі показники нанесення шкоди та витривалості.",
            "Epic": "Предмет має Епічну якість, найвищі характеристики та бонуси."
        }
    
    def __init__(self, name: str, rarity: str, damage: int, vitality: int, bonus: list = []) -> None:
        """Це Конструктор, він ініціалізує Обєкт за допомогою Аргументів які передаються в нього та присвоюються Атрибутам
        
        :param name: Цей аргумент визначає назву Меча;
        :param rarity: Цей аргумент відповідає за рідкісність предмету Меча;
        :param damage: Характеристика нанесення шкоди нашим Мечем;
        :param vitality: Характеристика міцності Меча;
        :param bonus: Список додаткових бонусів для Меча (Не реалізований функціонал);
        """
        self.name = name
        self.rarity = rarity if rarity in list(SwordsSecond.rarity_map.keys()) else None
        self.damage = damage
        self.vitality = vitality
        self.bonus = bonus
        self.buff: list = []
    
    # Створимо альтернативний конструктор, який буде створювати предмет меча відповідно до заданої рідкісності
    @classmethod
    def create_with_rarity(cls, name: str, rarity: str):
        """Альтернативний конструктор, створює обєкт Меча відповідно до переданої рідкісності
        :param name: Імя меча 
        :param rarity: Рідкісність
        """
        l = list(SwordsSecond.rarity_map.keys())
        if rarity in l:
            i = l.index(rarity) # тут ми знайдемо під яким індексом знаходиться задана якість меча
            return cls(name, rarity, 3 + i, 5 + i**2)
        raise ValueError(f"Введено направильне значення рідкісності предмету, повинно бути: {l}")
    
    @classmethod
    def create_with_random_rarity(cls, name: str):
        """Альтернативний конструктор, створює обєкт Меча з випадковою рідкісністю
        :param name: Імя меча
        """
        return cls.create_with_rarity(name, choice(list(SwordsSecond.rarity_map.keys())))

    @property
    def hit(self):
        """Визначає кількісь шкоди яку ми можемо нанести супротивнику"""
        return self.damage
    
    @property
    def rarity_type(self):
        """Дана проперті буде нам надавати інформацію про рідкісність предмету та визначати чи відповідає вона умовам"""
        if self.rarity in SwordsSecond.rarity_map.keys():
            return SwordsSecond.rarity_map[self.rarity]
        return "Предмет зломаний, можете викинути його у сміття."

    @property
    def info(self) -> str:
        """Виводить інформацію про Меч.
        :param self: Нам потрібен лише вказівник на обєкт з яким ми працюємо; 
        """
        p = f"""<<< Характеристика предмету {self.name} >>>
Рідкісність: {self.rarity}, {self.rarity_type}
Нанесення Шкоди: {self.damage}
Витривалість: {self.vitality}
Унікальний ефект: {self.bonus}
Накладені Бафи: {self.buff}"""
        #print(p)
        return p
    
    def apply_random_buff(self) -> str:
        """Накладає випадковий Баф на меч. Бафи реалізовані як окремий клас."""
        # Вибираємо випадковий баф з можливих
        m = choice([method for method in dir(Buff) if method.startswith('__') is False])
        # застосовуємо випадковий баф до меча, але не викликаємо його
        bb = Buff(self)
        b = getattr(bb, m) # тут в нас повертається функція
        # викликаємо власне вибраний метод для накладання бафу
        b() # тут викличеться функція та будуть змінені характеристики меча
        self.buff.append(bb)
        return f"Наклали баф {bb.name} що додає Шкоди: {bb.damage} та Витривалості: {bb.vitality} на меч {self.name}"
        #return "Неправильний баф, нічого не накладено"
    
    def expire_buff(self):
        """Завершує дію накледених бафів"""
        if self.buff:
            b = self.buff.pop()
            self.damage -= b.damage
            self.vitality -= b.vitality
            return f"Завершилась дія бафу {b.name}"
        return "На мечі немає ніяких бафів"
    
    def attack(self, item):
        if isinstance(item, SwordsSecond):
            item.vitality -= self.damage
            return f"{self.name} замахнувся та наніс {self.damage} шкоди мечу {item.name} у якого залишилось витривалосьті {item.vitality}"
        else:
            return f"{self.name} промахнулись та попали НЕ по мечу а по {item}!"

        

# тут ми тестуємо всі доступні нам спрособи створення обєктів за допомогою основного конструктора або альтернативних
s1 = SwordsSecond("Простий меч", "Green", 4, 8, ["Зцілення +1"])
print(s1.info)

s2 = SwordsSecond.create_with_rarity("Молот Богів", "Epic")
print(s2.info)

s3 = SwordsSecond.create_with_random_rarity("Зоряний Рапір")
print(s3.info)
```

- Умови гри: потрібно мати 2 гравці, які по черзі (по кроково) будуть вводити відповіді з клавіатури та робити певні дії на підсилення Меча або його атака і тд;

```python
from random import shuffle

swords_names = [
"Меч Смертівника", 
"Драконобійчий Клинок", 
"Стрілецька Сага", 
"Кований Піднебінник",
"Тіньовий Катана",
"Ластівчиний Клинок",
"Зоряний Рапір",
"Проклятий Кинджал",
"Молот Богів",
"Легендарний Лезо"]


player1 = str(input("Введіть імя гравця №1: "))
s1 = SwordsSecond.create_with_random_rarity(f"{choice(swords_names)} яким володіє {player1}")
s1.player = player1 # тут ми задаємо динамічний атрибут, який буде вказувати якому гравцю належить даний меч
player2 = str(input("Введіть імя гравця №2: "))
s2 = SwordsSecond.create_with_random_rarity(f"{choice(swords_names)} яким володіє {player2}")
s2.player = player2

print(f"""
Гравець №1: {s1.player} володіє Мечем:\n{s1.info}
Гравець №2: {s2.player} володіє Мечем:\n{s2.info}
""")

players = [s1, s2] # згрупував Мечі гравціі у список
shuffle(players) # помішав список щоб випадково визначити хто буду ходити першим

for turn in range(5):
    print(f"{10*'<'} РОЗПОЧИНАЄМО ХІД №{turn} {10*'>'}")
    shuffle(players) # помішав список щоб випадково визначити хто буду ходити першим
    for p in players:
        print(f"Випав хід для гравця {p.player}")
        action = int(input("Введіть 1 щоб атакувати, 2 щоб накласти випадковий Баф: "))
        if action == 1:
            # спробуємо видали з списка себе
            pp = players.copy()
            pp.remove(p)
            pp.append("Камінь") # додаємо якийсь обєкт, який не є мечем але ми можемо по ньому вдарити
            print(p.attack(choice(pp)))
        elif action == 2:
            print(p.apply_random_buff())
        else:
            print("Введено неправильне значення, гравець пропускає хід.")
    print(f"""
    <+> Гравець №1: {s1.player} Стати після ходу №{turn}:\n{s1.info}
    <+> Гравець №2: {s2.player} Стати після ходу №{turn}:\n{s2.info}
    """)
    if s1.vitality <= 0 or s2.vitality <= 0:
        print("ГРА ЗАВЕРШИЛАСЬ, В ОДНОГО З ГРАВЦІВ ЗЛОМАВСЯ МЕЧ")
        print(f"Виграв гравець {s1.player}" if s1.vitality > s2.vitality else f"Виграв гравець {s2.player}")
        break
```
