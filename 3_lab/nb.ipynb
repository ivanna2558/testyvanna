# Розпочинаємо основи тестування
- тести створюються для перевірки правильності роботи основної програми;
- в своїй основі, тести це також програми, або частинка коду яка запускає на виконання основний код та перевіряє правильність його роботи;
- в основному ми будемо зосереджуватись на Юніт тестуванні (пееревірки роботи якось юніта або частини коду) або на Функціональному тестуванні (перевірки чи код працює так як задумувалось);
- всі тести будуть в окремому файлі (можна також розбивати тести на різні файли, і поміщати у окрему папку, але ми так робити не будемо);
- ми не зможемо напряму працювати з класом Меча, написаним у роботі 2, тому ми просто будемо копіювати код, модифікувати його якщо потрібно, та формувати у реальний проект, з використання тестів;
- деколи тестування сприймають як дослідження по принципу чорного ящика де ми знаємо які вихідні дані мають бути при заданих вхідних (і тому ми очікуємо що ми отримуємо та можемо це перевіряти) - така перевірка на очікування може вважатись тестуванням - а точніше - твердженням;
- твердження виконуються за допомогою конструкції assert яка за функціоналом дуже подібна до блоку if, давайте спробуємо це на практиці:

```python
# Задомо собі різні типи даних з якими будемо працювати
a = 1
b = "1"
c = '1'
d = "one"
e = 1.0

# застосовуємо конструкцію тверджень
assert a >= 0, f'Твердження не є істинним, {a} не є меншим за 0!'

assert isinstance(b, str), f"Значення повинно бути {str}, а отримуємо {type(b)}"

# це твердження буде видавати неістину бо цифра не є рівною символу
#assert a == b, f"Дві величини не є однаковими, {a} не рівне {b}"

# помилкаа бо перевірка is співставляє типи які порівнюються
#assert a is e, f"Помилка твердження, {a} не є тотожним {e}, {type(a)} не відповідає {type(e)}"

assert len(d) >= 3, "Слово повинно мати більше 3 букв!"
```

-  все що ми робимо за допомогою assert можна рганізувати конструкцією if однак ця конструкція буде займати дещо більше коду;
- перепишемо останнє твердження за допомогою if;

```python
d = "one"
if len(d) >= 3:
    pass #ми нічого не робимо бо твердження є істинним
else:
    raise AssertionError("Твердження не є істинним у блоці if, слово повинно мати більше 3 букв!")

# можемо трохи хменшити код але потрібно інвертувати умову
d = "on"
if not len(d) >= 3:
    raise AssertionError("Твердження не виконалось, слово повинно мати більше 3 букв!")

# можна це ще більше спростити написавши if в одну стрічку
# в нас не виходить переписати
# ось результат ChatGPT
assert len(d) >= 3, "Твердження не виконалось, слово повинно мати більше 3 букв!" if not len(d) >= 3 else None

# як бачимо, застосування assert нам дозволить написати if в один рядочок, і варіант від ChatGPT може навіть обійтись без блоку if
```

- ми не обмежені в кількості тверджень які можемо використовувати при написанні тестів або при перевірці коду;
- якщо всі твердження будуть вірними, то і тестований код працює правильно;
- при застосуванні тверджень, воснвному перевіряються характеристики досліджуваного предмету;
- дані твердження можна уявити собі як критерії до чогось, якщо критерії задовільняються, все вірно - і ми протестували наш код.

```python
# перевіряємо стрічку записану у змінну
for a in ["Фраза для Тесту", "Тестова фраза", "Тестування "]:
    assert isinstance(a, str)
    assert len(a) >= 5
    assert "Тест" in a, "Фраза повинна містити слово 'Тест'."
    assert ' ' in a, "Фраза повинна складатись з декількох слів!"
    print(f"Всі твердження були вірними, тому наша фраза є вірною: {a}")
```
